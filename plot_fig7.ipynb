## Notebook for Fig. 7 of the paper

Here, we compare our 7-parameter model with an alternative model ("s4") that employs chiral EFT only up to 1.5 n<sub>sat</sub>

```python
import numpy as np
import corner
import matplotlib.pyplot as plt
import pickle
import scipy 
from tqdm import tqdm
import os
```

```python
from matplotlib import cm, colors
import matplotlib
plt.rcParams['text.usetex'] = True

fontsize = 16
```

```python
bins = 35

c3 = np.loadtxt('Injection_results/s2_hierarchical.dat')[:,1]
n1, bins1, _ = plt.hist(c3,bins = bins,color='C0',alpha = 0.75 ,density=True,label=r'$2 n_\textrm{sat}$'+r'$(20 \  \textrm{events})$')

c3 = np.loadtxt('Injection_results/s4_hierarchical.dat')[:,1]
n2, bins2, _ = plt.hist(c3,bins = bins,color='C0',alpha=0.75,histtype='step',density=True,label=r'$1.5 n_\textrm{sat}$'+r'$(20 \  \textrm{events})$')

c3 = np.loadtxt('Injection_results/s4_hierarchical_50.dat')[:,1]
n3, bins3, _ = plt.hist(c3,bins = bins,color='C1',histtype='step',density=True,label=r'$1.5 n_\textrm{sat}$'+r'$(50 \  \textrm{events})$')

c3 = np.loadtxt('Injection_results/s4_hierarchical_100.dat')[:,1]
n4, bins4, _ = plt.hist(c3,bins = bins,color='C4',histtype='step',density=True,label=r'$1.5 n_\textrm{sat}$'+r'$(100 \  \textrm{events})$')

c3 = np.loadtxt('Injection_results/s4_hierarchical_400.dat')[:,1]
n5, bins5, _ = plt.hist(c3,bins = bins,color='C3',histtype='step',density=True,label=r'$1.5 n_\textrm{sat}$'+r'$(400 \  \textrm{events})$')

plt.plot( np.linspace(-3.61*2,0), 1./(3.61*2)*np.ones(50),color='maroon',alpha = 0.5,zorder=1,label=r'$\textrm{Prior}$')
plt.plot( [-3.61*2,-3.61*2] , [0,1./(3.61*2)] ,color='maroon' ,alpha = 0.5,zorder=1)
plt.plot( [0,0] , [0,1./(3.61*2)] ,color='maroon' ,alpha = 0.5,zorder=1)

plt.minorticks_on()

plt.tick_params(axis='both',which='both',direction='in',bottom=True,top=True,left=True,right=True,labelsize=fontsize-1)

plt.legend(fontsize=fontsize-1,frameon=False)

plt.xlim(right=0.4,left = -3.7*2)

plt.xlabel('$c_3$'+r' $\textrm{[GeV}^{-1}\textrm{]}$',fontsize=fontsize)
plt.ylabel(r'$\textrm{PDF}$',fontsize=fontsize)

plt.savefig('figures/15nsat.pdf',bbox_inches='tight')
```
